```python
import pandas as pd
import numpy as np
```

### Dyskretyzacja

```python
np.random.seed(0)
age = np.random.randint(0, 106, 40)
age
```

```python
df = pd.DataFrame(age, columns=["Wiek"])
df
```

```python
# dyskretyzacja
df["Kategoria"] = pd.cut(df.Wiek, bins=[0,18,65,110], right=False,labels=["Dzieci", "Dorośli","Seniorzy"])
df
```

```python
np.random.seed(0)
data = np.random.normal(5, 1, 5000)
df = pd.DataFrame(data,columns=["x"])
df.head(10)
```

```python
df.describe()
```

```python
df["kwartyl"] = pd.qcut(df.x, 4, labels=[1,2,3,4])
df.head(20)
```

### One-hot encoding

```python
np.random.seed(0)
data = np.random.choice(["Pon", "Wto", "Śro"], 20)
df = pd.DataFrame(data, columns=["DT"])
df
```

```python
pd.concat([df, pd.get_dummies(df)], axis=1)
```
