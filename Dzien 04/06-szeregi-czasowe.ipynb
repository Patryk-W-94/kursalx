```python
df.groupby("Year")["Zamkniecie"].mean()
```

```python
df["Quarter"] = df.Data.dt.quarter.astype('object')
df.groupby(["Year","Quarter"])["Zamkniecie"].mean()
```

```python
df.Zamkniecie.plot(color="red", legend=True)
line2 = df.Zamkniecie.rolling(45).mean()
line2.name = "MA45"
line2.plot(figsize=(16,6), grid=True, color="green", legend=True)

line3 = df.Zamkniecie.rolling(90).mean()
line3.name = "MA90"
line3.plot(figsize=(16,6), grid=True, color="yellow", legend=True)
```

```python
df.Zamkniecie.rolling(30).mean().plot()
```

### Wstega Bollingera

```python
WINDOW_SIZE = 40
RATIO = 2.5
df["LowerRibbon"] = df.Zamkniecie.rolling(WINDOW_SIZE).mean() - RATIO*df.Zamkniecie.rolling(WINDOW_SIZE).std()
df["UpperRibbon"] = df.Zamkniecie.rolling(WINDOW_SIZE).mean() + RATIO*df.Zamkniecie.rolling(WINDOW_SIZE).std()
df
```

```python
import matplotlib.pyplot as plt
```

```python
df3 = df.loc["2021", ["Zamkniecie", "LowerRibbon", "UpperRibbon"] ]
df3.plot(figsize=(16,12), grid=True)
plt.fill_between( df3.index, df3.LowerRibbon, df3.UpperRibbon, color="red", alpha=0.2 )
```
